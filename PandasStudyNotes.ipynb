```python
#Pipe Functions - allows you to apply a function to the dataframe - chaining opearations

data = {'one'  : pd.Series([1, 2, 3, 4]),
        'two'  : pd.Series([10, 20, 30, 40]),
        'three': pd.Series([100, 200, 300, 400]),
        'four' : pd.Series([1000, 2000, 3000, 4000])}
df = pd.DataFrame(data)
df
```

```python
#add 
def add_(i,j):
    return i+j

print(df.pipe(add_,10))

print('-'*25)


#subtract
def subtract_(i,j):
    return i-j

print(df.pipe(subtract_,2))

print('-'*25)
```

```python
#mean squared

def mean_(col):
    return col.mean()

def square(i):
    return (i**2)

df.pipe(mean_).pipe(square)
```

```python
#Groupby function - used to split a dataframe into groups based on one or more columns

cricket = {'Team'   : ['India', 'India', 'Australia', 'Australia', 'SA', 'SA', 'SA', 'SA', 'NZ', 'NZ', 'NZ', 'India'],
           'Rank'   : [2, 3, 1,2, 3,4 ,1 ,1,2 , 4,1,2],
           'Year'   : [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
           'Points' : [876,801,891,815,776,784,834,824,758,691,883,782]}

df = pd.DataFrame(cricket)
print (df)

print("-"*50)

#group by teams 
df.groupby('Team').groups
```

```python
#group by team and year and then return the specific team and year from that 
print(df.groupby(['Team','Year']).get_group(('Australia',2014)))

print("-"*50)

#if the data is not present , we will get an error
#df.groupby(['Team','Year']).get_group(('USA',2014))

#total points teams have scored
print(df.groupby('Team').sum()['Points'])

print("-"*50)

#total points teams have scored - sort descending
print(df.groupby('Team').sum()['Points'].sort_values(ascending = False))

print("-"*50)

#mean points teams have scored
print(df.groupby('Team').mean()['Points'])
```

```python
groups= df.groupby('Team')


#as it is a series - we can run aggregate functions on it
groups['Points'].agg([np.sum,np.mean,np.max,np.min])
```

```python
#Filter function can be used with groupby 

#team SA appears 4 times
print(df.groupby('Team').filter(lambda x: len(x)==4))

print('-'*50)

#team NZ andIndia appears 3 times
print(df.groupby('Team').filter(lambda x: len(x)==3))
```
