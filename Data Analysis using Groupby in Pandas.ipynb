```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

df = pd.read_csv("googleplaystore.csv")
df.head()
```

```python
impute = SimpleImputer(missing_values = np.nan , strategy = "mean")
impute.fit(df.iloc[: , 2:3].values)
df.iloc[: , 2:3] = impute.transform(df.iloc[: , 2:3].values)
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df
```

### Name the apps that are of art and Design category that have rating greater than 4.5 in an ascending order(wrt rating)

```python
#sort values - sort function in pandas
df_pr = df[df["Category"] =="ART_AND_DESIGN"]
df_pr = df_pr[df_pr["Rating"] > 4.5]

df_pr.sort_values( by = "Rating" , ascending = True)
```

```python
#descending order :
df_pr.sort_values( by = "Rating" , ascending = False)
```

### Name the apps that are of Art and Design category that have rating greater than 4.5 in an ascending order(wrt reviews)

```python
#sort values - sort function in pandas
df.head(10)
df_pr = df[df["Category"] =="ART_AND_DESIGN"]
df_pr = df_pr[df_pr["Rating"] > 4.5]

#did not work as Review is a string
df_pr.sort_values( by = "Reviews" , ascending = False)
```

#### Which category is having max average rating

```python
#Group wrt Category, find mean of Rating
df.groupby("Category").mean()["Rating"].sort_values()
```

### How many apps are there in each category in Descending order?

```python
df.groupby("Category").count()["Type"].sort_values(ascending = False)
```

### How many free apps are there in each category in Descending order?
