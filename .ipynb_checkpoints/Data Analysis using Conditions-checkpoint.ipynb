```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

df = pd.read_csv("googleplaystore.csv")
df.head()
```

```python
impute = SimpleImputer(missing_values = np.nan , strategy = "mean")
impute.fit(df.iloc[: , 2:3].values)
df.iloc[: , 2:3] = impute.transform(df.iloc[: , 2:3].values)
df
```

```python
df.dropna(inplace= True)
```

```python
df.isnull().sum()
```

### How many apps are there in ART_AND_DESIGN with rating > 4.5?

```python
df_pr = df[df['Category'] =="ART_AND_DESIGN"]
len(df_pr[df_pr["Rating"] > 4.5])
```

### How many free apps are there in FAMILY with rating > 4.5?

```python
df_pr = df[df['Category'] =="FAMILY"]
df_pr = df_pr[df_pr["Rating"] > 4.5]
len(df_pr[df_pr["Type"] == "Free"])
```
