```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
sns.countplot(x ="survived" , data =df)
```

#### Bar Plot

```python
sns.countplot(x ="survived" ,hue ="class", data =df)
```

#### HeatMap  - use if you have more category values in the category 

```python
#First create pivot table
#enter column names with the pandas function crosstab
x = pd.crosstab(df["survived"],df["class"])
x
```

```python
sns.heatmap(x)
```

```python
#format --> fmt --> round of to 1 decimal point
#annotation --> annot --> to have the label for values on the heat map
sns.heatmap(x, annot = True , fmt = ".1f")
```

```python
#format --> fmt --> round of to 1 decimal point
#annotation --> annot --> to have the label for values on the heat map
#colourmap --> cmap --> to change the colour of the heat map
sns.heatmap(x, annot = True , fmt = ".1f", cmap="coolwarm")
```

#### Strip Plot

```python
sns.stripplot(data =df, x ="class" , y ="age" , hue ="sex")
```
