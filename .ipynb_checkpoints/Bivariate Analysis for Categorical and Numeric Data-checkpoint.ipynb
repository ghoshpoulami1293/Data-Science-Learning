```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

#### Bar Graph

```python
sns.barplot(data = df , x ="class", y = "age" )
```

```python
sns.barplot(data = df , x ="class", y = "age", estimator = "sum")
```

```python
sns.barplot(data = df , x ="class", y = "fare", estimator = "sum")
```

```python
sns.barplot(data = df , x ="class", y = "age", hue = "sex")
```

```python
sns.barplot(data= df , x ="class", y = "fare", hue ="sex" , estimator ="sum")
```

#### Box Plot

```python
sns.boxplot(data=df, x ="class" ,y ="age")
```

```python
#same1
sns.boxplot(data=df, x ="class", hue ="sex",y ="age")
```

```python
#same2
sns.boxplot(data=df, x ="class",y ="age" , hue ="sex")
```

```python
sns.boxplot(data=df, x ="class",y ="age" , hue ="survived")
```

#### Violin Plot

```python
sns.violinplot(data=df, x ="class",y ="age")
```

```python
sns.violinplot(data=df, x ="class",y ="age" , hue ="survived")
```

#### Plotting Probability Distribution using distplot

```python
sns.distplot(df[df["class"] =="Second"]["age"])
```
