### <font color = "green">Linear Regression</font>

#### Forward Propagation

f(x) = m*x +c

```python
#Function to define forward propagation:

def forward_propagation(train_input, parameters):
    m = parameters["m"]
    c = parameters["c"]
    
    #x is a numpy array, so to multiply x with the vector m :
    predictions = np.multiply(m, train_input) +c
    return predictions
```

#### Cost function

summation(1/2n * (y - f(x)^2))

```python
def cost_function(predictions, train_output):
    cost = np.mean((train_output - predictions)**2)*0.5
    return cost
```

#### Gradient Descent for Backpropagation

df = summation(f(x) - y)/n

dm = df * x

dc = df * 1

 ##### d(cost)/dm = d(cost)/df * df/dm

```python
def backward_propagation(train_input, train_output, predictions):
    
    derivatives=dict()
    
    df = predictions - train_output
    
    dm = np.mean(np.multiply(df, train_input))
    
    dc = np.mean(df)
    
    derivatives["dm"]= dm 
    derivatives["dc"]= dc
    
    return derivatives
```

#### Update Parameters

m = m - (learning_rate * dm)

c = c - (learning_rate * dc)

```python
def update_parameters(parameters , derivatives, learning_rate):
    parameters["m"] = parameters["m"] - learning_rate * derivatives["dm"]
    parameters["c"] = parameters["c"] - learning_rate * derivatives["dc"]
    
    return parameters
```

#### Train Model

```python
def train(train_input, train_output, learning_rate, iters):
    #random parameters
    parameters=dict()
    parameters["m"] = np.random.uniform(0,1) # any value between 0 to 1, this is a uniform distribution
    parameters["c"] = np.random.uniform(0,1) # any value between 0 to 1, this is a uniform distribution
    
    plt.figure()
    
    #loss -- to see how the cost function is reducing everytime
    loss = list()
    
    #iteration
    for i in range(iters):
        
        #forward propagation
        predictions = forward_propagation(train_input, parameters)
        
        #calculate cost
        cost = cost_function(predictions, train_output)
        
        loss.append(cost)
        #print loss per iteration
        print(f"Iteration ={i+1} , Loss ={cost}")
        
        #plot the function as well
        
        #define figure and axes 
        fig, ax = plt.subplots()
        
        #for each axis:
        #plot input vs output plot - trianing data
        ax.plot(train_input, train_output, "+" , label = "Original Data")
        
        #plot input vs output plot - our model's hypothesis
        ax.plot(train_input, predictions, "*" , label = "Original Data")
            
        #Legend - label for each and every plot that is being made
        legend = ax.legend()
        
        #then the actual plot
        ax.plot(train_input, train_output, "+" , label = "Original Data")
        ax.plot(train_input, predictions, "*" , label = "Original Data")
        
        plt.show()        
        
        #backward_propagation
        derivatives = backward_propagation(train_input, train_output, predictions)
        
        #update parameters 
        parameters = update_parameters(parameters, derivatives, learning_rate)
        
    return parameters,loss
```

#### Training

```python
#very low learning rate
#keep iterations = 20
parameters , loss = train(train_input, train_output, 0.0001,20)
```
