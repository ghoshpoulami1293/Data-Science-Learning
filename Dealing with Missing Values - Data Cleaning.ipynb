### Filling  the missing values using imputation: Numerical

```python
df3.head()
```

```python
df3.isna().sum()
```

```python
#Using mean function
mean_value = df3["reviews_per_month"].mean()
mean_value
```

```python
fill using basic functions, not SimpleImputer
df3['reviews_per_month'].fillna(mean_value,inplace = True)
df3['reviews_per_month']
```

```python
df3.isna().sum()
```

### Filling  the missing Categorical values

```python
df3.nunique()
```

```python
#value_counts - sorted values such that the date that occurs max times is at index 0 --> equivalent to mode function
df3["last_review"].fillna(df3["last_review"].value_counts().index[0],inplace=True)
```

```python
df3['last_review']
```

```python
df3.isna().sum()
```

```python
df4.isna().sum()
```

```python
#Replace with a custom value
df4["last_review"].fillna("Not Reviewed" , inplace = True)
```

```python
df4.sample(10)
```

```python
df4.isna().sum()
```

```python
#use interpolate function to fill in the missing values in the dataset using the surrounding values 
#(finds the mean of the surrounding values and replaces the missing value with this)

df5["reviews_per_month"].interpolate(inplace=True)
```
